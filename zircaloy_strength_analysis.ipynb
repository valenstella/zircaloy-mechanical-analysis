```python
import pandas as pd

# Crear tabla
df = pd.DataFrame({
    "Temp": [25, 100, 200, 300, 400],
    "Yield_Strength": [480, 430, 380, 320, 240]
})

# Mostrar tabla
print(df)

# Guardar CSV
df.to_csv("zircaloy_strength.csv", index=False)

print("Archivo guardado correctamente")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("zircaloy_strength.csv")

print(df.describe())

#plt.plot(df["Temp"], df["Yield_Strength"])
plt.scatter(df["Temp"], df["Yield_Strength"],color="r")
plt.xlabel("Temperature (°C)")
plt.ylabel("Yield Strength (MPa)")
plt.title("Zircaloy-4 Yield Strength vs Temperature")
plt.grid()
plt.show()
```

```python
import numpy as np

coef = np.polyfit(df["Temp"],df["Yield_Strength"],2)

poly = np.poly1d(coef)

x = np.linspace(0,400,100)

plt.scatter( df["Temp"],df["Yield_Strength"],color="r")

plt.plot(x, poly(x))
plt.savefig("zircaloy_strength.png", dpi=300)
plt.show()
```

```python
print("Analysis completed successfully")
print(f"Maximum strength: {df['Yield_Strength'].max()} MPa")
print(f"Minimum strength: {df['Yield_Strength'].min()} MPa")
```
